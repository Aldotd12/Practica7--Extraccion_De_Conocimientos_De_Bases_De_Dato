# Importando API

```python
# Librerias necesarias
import pandas as pd
import requests
```

```python
# Obtenemos los datos 
url = "https://api.covidtracking.com/v1/us/daily.json"  
response = requests.get(url)
comentarios = response.json()
```

```python
df_comentarios = pd.DataFrame(comentarios)
```

```python
df_comentarios.head()
```

```python
df_comentarios.info()
df_comentarios.describe()
```

# Limpiar Datos

```python
# Elimino todos los json que tengan states = 56.000000
df_filtrado = df_comentarios[df_comentarios["states"] != 56.000000].copy()
```

```python
# hago una consulta para ver el filtrado
df_filtrado.head()
```

```python
# Muestro metadatos actuales
df_filtrado.info()
df_filtrado.describe()
```

```python
# SE transforman todos los states a id 2 a 99
# Muestro los primeros registros del dataframe
df_filtrado['states'].head()
```

```python
#Remplazo de los datos
df_filtrado.loc[:, 'states'] = df_filtrado['states'].replace(51, 'Total')
```

```python
#ejecutamos el comando para ver si se reemplazo
df_filtrado['states'].head() 
```

```python
# 4.1 – Guardar a CSV
df_covid = pd.DataFrame(comentarios)

# 3. Guardar en archivo CSV
df_covid.to_csv("covid_data.csv", index=False)
print("Archivo 'covid_data.csv' generado")

# 4. Conexión a MySQL (ajusta los valores con los de tu entorno)
engine = create_engine("mysql+pymysql://root:aldo@localhost/covidPractica8")

# 5. Guardar en MySQL como tabla
df_covid.to_sql(
    name='covid_us_daily',
    con=engine,
    if_exists='replace',
    index=False
)
print("Tabla 'covid_us_daily' creada en MySQL")
```
